# Finding out the Missing Values

```python
Data.isnull().sum()
```

# Imputing the missing value

```python
Data['city'].fillna(value='Unknown',inplace=True)
Data['Killed'].fillna(Data['Killed'].mean(),inplace=True)
Data['Wounded'].fillna(Data['Wounded'].mean(),inplace=True)
Data['Target'].fillna(value='unknown',inplace=True)
Data['Summary'].fillna(value='nothing',inplace=True)
Data['Motive'].fillna(value='nothing',inplace=True)
```

```python
Data.isnull().sum()
```

# Data Visulization of the Global Visulization

```python
plt.figure(figsize=(15,6))
sns.countplot(Data['Year']).set_title('Year wise attacks')
plt.xticks(rotation=90)
```

More Nuber of attack in 2014

```python
Data['Region'].value_counts()
```

```python
plt.figure(figsize=(15,6))
sns.countplot(Data['Region']).set_title('Region Wise attack')
plt.xticks(rotation=60)
```

Middle east and north africa is most attacked region

```python
a3 = Data.groupby(by='AttackType').agg(['count'])
a4 = a3['Year']
a4.columns = ['count']
a4 = a4.reset_index()
a4
```

```python
plt.figure(figsize=(15,10))
sns.barplot(y = 'AttackType', x = 'count', data = a4)
plt.title("The Most Common Attack Types")
plt.show()
```

The most attacked type is Boming and explosion
